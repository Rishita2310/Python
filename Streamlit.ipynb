# Streamlit
- A python based framwork used to built interactive web pages for data science

```python
%%writefile first.py
import streamlit as st
st.set_page_config(page_title="Hello Streamlit",layout='centered')

st.title("Welcome to streamlit")
st.header("This is header")
st.subheader("This is a subheader")

st.text("This is st.text")
st.write("This is st.write")
st.markdown("This is `st.markdown`")

code="""
def add(a,b):
    return a+b
print(add(5,6))
"""

########## Sidebar,columns and expanders

st.sidebar.header("Profile settings")
name=st.sidebar.text_input("Faculty name")
dept=st.sidebar.selectbox("Department",['CE','CST','IT'])
col1,col2=st.columns([1,2])
with col1:
    st.write("Basic Info")
    st.write(name,dept)
with col2:
    st.subheader("About")
with st.expander("Subjects taught"):
    st.write("Python,Fsd,DE")

########## FORMS AND INPUT ##########
name = st.text_input("Enter name: ")
age = st.number_input("Age",0,80,18)
rating = st.slider("Rating",1,10,5)

if name:
    st.write(f"Hello {name}")

########## SELECTION WIDGETS ##########
course = st.selectbox("Course",['Python','FSD','DE'])
days = st.multiselect("Preferred Days",["MON","TUE","THUR"])
lec = st.radio("Lecture No", [1,2,3])

##########DATE TIME AND FILE UPLOD
from datetime import date,time
examdate=st.date_input("Exam Date",date.today())
starttime=st.time_input("Start time",time(14,0))
file=st.file_uploader("Upload CSV",type=['csv'])

######### MEDIA DISPLAY ########
st.subheader("Image Display")
st.image("http://picsum.photos/600/300",caption='Random image',use_container_width=True)

st.subheader("Local Image Display")
st.image("Flower.png",caption='local image',use_container_width=True)

st.subheader("Audio Example")
st.audio("Chal chhaiya chhaiya.mp3")

# st.subheader("Video Example")
# st.video("VID20251205133625.mp4")

## Creating and displaying data
import pandas as pd
data={
    "Student":['A','B','C','D'],
    "Marks":[85,90,95,20],
    "Passed":[True,True,True,False]
}
df = pd.DataFrame(data)

##Display dataframe
st.header('st.dataframe')
st.dataframe(df)

## Display table
st.header("static table")
st.table(df)

## json-javascript object notation
st.header("Json data")
st.json(data)

############ STATUS AND RUNNING TASKS ##########
st.info("Useful Information")
import time

if st.button("start long run"):
    progress=st.progress(0)
    with st.spinner("processing"):
        for i in range(100):
            time.sleep(0.03)
            progress.progress(i+1)
    st.success("Task completed.")

```
