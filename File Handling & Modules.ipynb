```python
l=list()
d=dict()
handle=open("Downloads/mbox-short.txt",encoding='utf-8')
for line in handle:
    line=line.rstrip()
    if not line.startswith("From "): continue
    words=line.split()
    l.append(words[1])
for i in l:
    d[i]=d.get(i,0)+1
print(d)
b=max(d.values())
for k,v in d.items():
    print(k,v)
handle.close()
print(b)
```

```python
def pager(f):
    with open(f, encoding="utf-8") as file:
        lines = file.readlines()
        total = len(lines)
        start = 0
        end = min(25, total)

        while start < total:
            for i in range(start, end):
                print(lines[i], end="")

            user = input("continue or stop: ")
            if user.lower() == "stop":
                break

            start += 25
            end = min(start + 25, total)

print(pager("alice.txt"))
```
