```python
l1=soup.find_all("div",class_="student")[0:2]
print(len(l1))
```

```python
name=[]
dep=[]
email=[]
ext=[]
social=[]
link=[]
img=[]
for i in l1:
    n=i.find("h2").text
    name.append(n)
    d=i.find("p").text
    dep.append(d)
    e=i.find("span").text
    ex=e.split('@')
    email.append(ex[0])
    ext.append(ex[1])
    s=i.find("a").text.strip()
    social.append(s)
    l=i.find("a")
    link.append(l["href"])
    im=i.find("img")
    img.append(im["src"])
    
```

```python
data={"Title":title,
     "Name":name,
     "Department":dep,
     "Email":email,
      "Ext":ext,
     "Social":social,
     "Link":link,
     "Image":img}
```

```python
df=pd.DataFrame(data)
df
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url="https://www.politifact.com/factchecks/list/"
data=requests.get(url)
data.status_code
html=data.text
```

```python
soup=BeautifulSoup(html,"html.parser")
z=soup.find_all("li",class_="o-listicle__item")
name=[]
date=[]
info=[]
author=[]
for i in z:
    n=i.find("a",class_="m-statement__name").text.strip()
    name.append(n)
    d=i.find("footer",class_="m-statement__footer").text.strip()
    a=d.split("•")
    date.append(a[1])
    author.append(a[0])
    inf=i.find("div",class_="m-statement__desc").text.strip()
    info.append(inf)
```

```python
import pandas as pd
data={
    "Name":name,
    "Info":info,
    "Date":date,
    "Author":author
}
df=pd.DataFrame(data)
```

```python
with open("flipkart.html","r",encoding="UTF-8") as f:
    html=f.read()
```

```python
soup=BeautifulSoup(html,"html.parser")
data=soup.find_all("div",class_="_3pLy-c row")
name=[]
rating=[]
Ram=[]
battery=[]
og=[]
dis=[]
for i in data:
    n=i.find("div",class_="_4rR01T").text.strip()
    name.append(n)
    r=i.find("div",class_="_3LWZlK").text.strip()
    rating.append(r)
    ra=i.find("li",class_="rgWa7D").text.strip()
    Ram.append(ra)
    ba=i.find_all("li",class_="rgWa7D")[3]
    battery.append(ba.text)
    o=i.find("div",class_="_3I9_wc _27UcVY")
    if o:
        og.append(o.text)
    else:
        og.append(0)
    d=i.find("div",class_="_30jeq3 _1_WHN1")
    if d:
        dis.append(d.text)
    else:
        dis.append(0)
```

```python
import pandas as pd
d={
    "Name":name,
    "Rating":rating,
    "RAM":Ram,
    "Battery":battery,
    "Original price":og,
    "Discount price":dis,
}
df=pd.DataFrame(d)
df
```

```python
o=[]
d=[]
for i in df["Original price"]:
    if i:
        o.append(i[1:].replace(",",""))
    else:
        o.append(0)
for i in df["Discount price"]:
    d.append(i[1:].replace(",",""))
```
