## Inheritance
- A class can be inherit from the other class
- It enables code reuse
- Constuctor, methods and attributes are inherited to the child class
- The parent class has no access to the child class
- Child class can override the methods of parent class. This is called overriding
- 'super()' mthod is used to invoke the parent class method and constructor

### Single Inheritance

````python
class Person:
    def __init__(self,name):
        self.name=name
    def show(self):
        print(self.name)
````

````python
class Professor(Person):
    pass
````

````python
p=Professor("ABC")
````

````python
p.show()
````

````python
p.name
````

### Multilevel Inheritance

````python
class P():
    def __init__(self,name):
        self.name=name
    def show(self):
        print(self.name)
class C1(P):
    pass
class C2(C1):
    pass
x=C2("Rishita")
x.show()
````

### Multiple Inheritance

````python
class P1():
    def __init__(self,name):
        self.name=name
    def show1(self):
        print(self.name)
class P2():
    def show2(self):
        print(self.name)
class C(P1,P2):
    pass
x=C("Rishita")
x.show1()
x.show2()
````

### Hiearchical Inheritance

````python
class P():
    def __init__(self,name):
        self.name=name
    def show(self):
        print(self.name)
class C1(P):
    pass
class C2(P):
    pass
x=C1("Rishita")
x.show()
y=C2("Rishita")
y.show()
````

### Constructor and method overriding

````python
class Father:
    def __init__(self):
        self.money=50000
        print("Father Constructor")
    def show(self):
        print(self.money)
class Son(Father):
    def __init__(self):
        super()
        self.money=5000
        self.car="XUV"
        print("Son Constructor")
    def show(self):
        print(self.money)
````

````python
s=Son()
s.show()
````

````python
f=Father()
f.show()
f.car
````

### super() Method
- If we write constructor in both parent and child class, the parent class constructor is not available in child class
- super() method is used to invoke parent class constructor from the child class

````python
class Father:
    def __init__(self):
        self.money=50000
        print("Father Constructor")
    def show(self):
        print(self.money)
class Son(Father):
    def __init__(self):
        super()
        self.money=5000
        self.car="XUV"
        print("Son Constructor")
    def show(self):
        super().show()
        print(self.money)
````

````python
s=Son()
s.show()
````

### method overloading

````python
class Car:
    def move(self):
        print("Car is running")
    def stop(self):
        print("Car is stopped")
class Person:
    def move(self):
        print("Person is running")
    def stop(self):
        print("Person is stopped")
````

````python
c=Car()
p=Person()
c.move()
c.stop()
p.move()
p.stop()
````

## Abstract class
- a class derived from ABC class belonging to abc module is known as abstract class in python
  ```python
  from abc import ABC,abstract method
  ```
#### Rules of abstract class
- One cannot create object of abstract class
- It is not necessary to declare all methods abstract in abstract class
- Abstract class can have abstract and concreate methods
- If the method is abstract, then the class must be abstract
- abstract methods of abstract class must be defined in the child classes

````python
from abc import ABC,abstractmethod
````

````python
class Defence(ABC):
    def __init__(self):
        self.id=101
    @abstractmethod
    def area(self):
        pass
    def gun(self):
        print("The gun is desert Eagle.")
````

````python
class Army(Defence):
    def area(self):
        print("Area is on land ",self.id)
````

````python
a=Army()
a.gun()
a.area()
````

````python
import math
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
class Rectangle(Shape):
    def calculate_area(self,l,b):
        print("Area of rectangle is= ",(l*b))
class Circle(Shape):
    def calculate_area(self,r):
        print("Area of circle is= ",(math.pi*r*r))
r=Rectangle()
r.calculate_area(2,4)
c=Circle()
c.calculate_area(10)
````
