```python
import pandas as pd
df = pd.DataFrame({
    "Gender":["M","F","M","F","M","F"],
    "Dept":["IT","IT","HR","HR","Finance","Finance"],
    "Salary":[50000,-55000,40000,70000,80000,45000],
    "ex":["1","2","3","4","5","6"]
})
df
```

```python
fig=px.bar(df,x="Dept",y="Salary",barmode="relative",color="Gender")
fig.show()
```

```python
fig=px.bar(df,x="Dept",y="Salary",barmode="group",color="Gender",width=800,height=500,
          hover_data=["ex"],color_discrete_map={"M":"yellow","F":"black"})
fig.show()
```

```python
data={
    "h":[2,3,4,5,6,7],
    "marks":[40,50,60,65,70,80],
    "Gender":["M","F","M","F","M","F"],
    "name":[40,50,60,65,70,80],
    "D":["IT","HR","IT","Finance","HR","Finance"]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
fig=px.scatter(df,x="h",y="marks",color="Gender",title="Marksheet",symbol="Gender",size="marks",hover_data=["D","name"],)
fig.show()
```

```python
fig=px.scatter_3d(df,x="h",y="marks",z="Gender",color="Gender",title="Marksheet",symbol="Gender",size="marks",hover_data=["D","name"],)
fig.show()
```

```python
fig=px.line(df,x="h",y="Gender",markers=True)
fig.show()
```

```python
fig=px.pie(df,values="h",names="D")
fig.update_traces(pull=[0.1,0,0],textinfo="percent+label")
fig.show()
```

```python
g=nx.Graph()
g.add_node(1)
g.add_nodes_from([10,20,30,40,50])
g.add_edge(1,10)
g.add_edge(40,10)
g.add_edge(1,20)
g.add_edge(50,10)
g.add_edge(30,10)
g.add_edges_from([(40,1),(30,1)])
# g.clear_edges()
print(g.has_edge(40,20))


nx.draw_networkx(g,node_size=850,node_color="deeppink",width=4,edge_color="blue") 
```

```python
g=nx.DiGraph()
g.add_node(1)
g.add_nodes_from([10,20,30,40,50])
g.add_edge(1,10)
g.add_edge(40,10)
g.add_edge(1,20)
g.add_edge(50,10)
g.add_edge(30,10)
g.add_edges_from([(40,1),(30,1)])

nx.draw_networkx(g,node_size=850,node_color="red",width=4,edge_color="blue") 
```
