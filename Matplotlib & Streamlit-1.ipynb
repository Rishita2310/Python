#### Subplots

````python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("A sine-curve")

plt.subplot(1,2,2)
y1=np.cos(x)
plt.plot(x,y1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A cosine curve")

plt.show()
````

````python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("A sine-curve")

plt.subplot(1,2,2)
y1=np.cos(x)
plt.plot(x,y1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A cosine curve")

plt.tight_layout()
plt.show()
````

````python
plt.suptitle("Family of graph!")

plt.subplot(3,2,1)        #3 rows, 1 column, 1st phase
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("A sine-curve")

plt.subplot(3,2,2)        #3 rows, 1 column, 2nd phase
y1=np.cos(x)
plt.plot(x,y1)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A cosine curve")

plt.subplot(3,2,3)        #3 rows, 1 column, 3rd phase
y2=np.tan(x)
plt.plot(x,y2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A tan curve")

plt.subplot(3,2,4)
y3=np.tanh(x)
plt.plot(x,y3)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A tanh curve")

plt.subplot(3,2,5)
y4=np.sinh(x)
plt.plot(x,y4)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A sinh curve")

plt.subplot(3,2,6)
y5=np.cosh(x)
plt.plot(x,y5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("A cosh curve")

# plt.subplot(3,2,7)
# y3=np.tanh(x)
# plt.plot(x,y3)
# plt.xlabel("X")
# plt.ylabel("Y")
# plt.title("A tanh curve")    #give error

plt.tight_layout()
plt.show()
````

#### Scatter

````python
x=range(6)
y=list(i*10 for i in range(1,7))
plt.scatter(x,y,label='a scatter',s=200)  #s=scatter-size
plt.legend()
plt.show()
````

#### Bar-Plot-vertical

````python
x=range(6)
y=list(i*10 for i in range(1,7))
plt.bar(x,y,width=0.2)
plt.show()
````

#### Bar-plot-Horizontal

````python
x=range(6)
y=list(i*10 for i in range(1,7))
plt.barh(x,y,height=0.6)
plt.show()
````

### PIE-Chart

````python
labels=["A","B","C","D","E"]
sizes=[40,10,20,14,16]
colors=['beige','gold','lightcoral','lightskyblue','yellowgreen']

plt.pie(sizes,labels=labels,colors=colors,shadow=True,autopct="%1.0f%%",startangle=90)
plt.show()
````

````python
labels=["A","B","C","D","E"]
sizes=[40,10,20,14,16]
colors=['beige','gold','lightcoral','lightskyblue','yellowgreen']

plt.pie(sizes,labels=labels,colors=colors,shadow=True,autopct="%1.0f%%",startangle=90,explode=[0.1,0,0,1,0])
plt.show()
````

````python
labels=["A","B","C","D","E"]
sizes=[40,10,20,14,16]
colors=['beige','gold','lightcoral','lightskyblue','yellowgreen']

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    shadow=True,
    autopct="%1.0f%%",
    startangle=90,
    explode=[0.1,0,0,0,0]
)
plt.legend(loc="upper right")
plt.show()
````

### Area plot/fill_between and skyplot

````python
import numpy as np
import matplotlib.pyplot as plt
x=range(1,6)
y=[1,4,6,8,4]

plt.fill_between(x,y)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
x=range(1,6)
y=[1,4,6,8,4]

plt.fill_between(x,y)
plt.plot(x,y,linewidth=10,color='m')
plt.show()
````

### Stackplot

````python
import numpy as np
import matplotlib.pyplot as plt
x=range(1,6)
y1=[1,4,6,8,9]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]

plt.stackplot(x,y1,y2,y3,labels=['A','B','C'])
plt.legend()
plt.show()
````
