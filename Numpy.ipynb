## np.concatenate
- It joins two given array at given axis

```python
c=np.arange(6).reshape(2,3)
d=np.arange(6,12).reshape(2,3)
print(c)
print(d)
```

```python
np.concatenate((c,d))
```

```python
np.concatenate((c,d),axis=0)
```

```python
np.concatenate((c,d),axis=1)
```

```python
np.concatenate((c,d),axis=2)
```

## `np.array_split`
- it opposite to join,will split the given into given split
- Split conditions- For an array of length l that should be split
    into n sections, it returns l % n sub-arrays of size l//n + 1
    and the rest of size l//n.

```python
np.array_split(np.array([1,2,3,4,5,6,7]),3)
```

## `np.where`
- it returns the indices of the given array where the given condition is satisfied

```python
a=np.array([40,34,67,56,45,87,12,68])
```
