count upper case, lower case, space, digits, characters

```python
s="This is pythOn programming 100 tutorial"
u=l=d=space=0
for i in s:
    if i.isupper():
        u=u+1
    elif i.islower():
        l=l+1
    elif i.isdigit():
        d=d+1
    else:
        space=space+1
print(space,u,l,d)
```

```python
s="123Hello"
digits=sum=avg=0
for i in s:
    if i.isdigit():
        digits+=1
        i=int(i)
        sum+=i
avg=sum/digits
print(digits)
print(sum)
print(avg)
```

shifting

```python
def shift(n,s):
    strnum=str(n)
    digits=len(strnum)
    if s>digits:
        res=strnum[::-1]
    else:
        res=strnum[s:]+strnum[:s]
    return int(res)
shift(12345,3)
```
