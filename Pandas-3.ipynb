```python
data={"maths":[80,90,85,70,200,500,0]}
df=pd.DataFrame(data)
df
```

```python
q1=df["maths"].quantile(0.25)
q3=df["maths"].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
lower=q1-(iqr*1.5)
upper=q3+(iqr*1.5)
print(lower)
```

```python
print(upper)
```

```python
df[(df["maths"]>=lower) & (df["maths"]<=upper) ]
```

```python
import seaborn as sns
sns.boxplot(df["maths"])
plt.show()
```
