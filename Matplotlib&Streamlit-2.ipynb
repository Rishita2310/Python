## Area plot/fill_between and stackplot:

````python
import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,4,6,8,4]
plt.fill_between(x,y)
plt.plot(x,y,linewidth=5,color="m")
plt.show()
````

## Stackplot:

````python
import numpy as np
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y1=[1,4,6,8,9]
y2=[2,2,7,10,12]
y3=[2,8,5,10,6]

plt.stackplot(x,y1,y2,y3,labels=["A","B","C"])

plt.legend()


plt.show()
````
